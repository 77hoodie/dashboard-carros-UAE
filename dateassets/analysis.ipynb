```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('uae_used_cars_10k.csv')
```

```python
df.info()
```

```python
df.head(10)
```

```python
df.isna().sum()
```

```python
df["price_per_km"] = df["Price"] / df["Mileage"].replace(0, 1)
```

```python
fig1 = px.scatter(df, x="Mileage", y="Price", color="Body Type", hover_data=["Make", "Model", "Year"], title="Preço em relação à quilometragem do carro")
fig1.update_layout(height=400)
```

```python
fig2 = px.box(df, x="Body Type", y="Price", color="Body Type", title="Distribuição de preços por tipo de carroceria")
fig2.update_layout(height=400)
```

```python
fig3 = px.histogram(df, x="Year", nbins=20, title="Distribuição por ano de fabricação")
fig3.update_layout(height=300)
```

```python
preco_medio = df.groupby('Transmission')['Price'].mean().sort_values(ascending=False).reset_index()
fig4 = px.pie(preco_medio, names='Transmission', values='Price', title='Distribuição do preço médio por tipo de transmissão')
fig4.update_traces(textposition='inside', textinfo='percent+label')
```

```python
preco_ano = df.groupby("Year")["Price"].mean().reset_index()
fig5 = px.line(preco_ano, x="Year", y="Price", title="Preço médio por ano de fabricação")
fig5.update_layout(height=400)
```

```python
modelos_top = df["Model"].value_counts().head(10).reset_index()
modelos_top.columns = ["Model", "Count"]
pd.DataFrame(modelos_top)
```

```python
contagem_marcas = df['Make'].value_counts().reset_index()
contagem_marcas.columns = ['Make', 'Count']
fig6 = px.bar(contagem_marcas, x='Make', y='Count',title='Quantidade de veículos por marca', labels={'Count': 'Quantidade', 'Make': 'Marca'})
fig6.update_layout(height=500, xaxis_tickangle=-45)
```

```python
preco_medio = df.groupby('Make')['Price'].mean().sort_values(ascending=False).reset_index()
fig4 = px.bar(preco_medio, x='Make', y='Price', title='Preço médio por marca', labels={'Price': 'Preço médio', 'Make': 'Marca'})
fig4.update_layout(height=500, xaxis_tickangle=-45)
```

```python
preco_km_marca = df.groupby('Make')['price_per_km'].mean().sort_values(ascending=False).reset_index()
fig6 = px.bar(preco_km_marca, x='Make', y='price_per_km', title='Relação preço por quilometragem por marca', labels={'price_per_km': 'Preço por km', 'Make': 'Marca'})
fig6.update_layout(height=500, xaxis_tickangle=-45)
```

```python
preco_medio_carroceria = df.groupby("Body Type")["Price"].mean().sort_values(ascending=False).reset_index()
fig7 = px.bar(preco_medio_carroceria, x="Body Type", y="Price", color="Body Type", title="Preço médio por tipo de carroceria", labels={"Price": "Preço médio", "Body Type": "Tipo de carroceria"})
fig7.update_layout(height=500, xaxis_tickangle=-45)
```
